```python
# gold medals for each country.
gold_medal.Team.value_counts().reset_index(name = 'Medal').head(5)
```

```python
# plot total gold medals of different countries.
totalgoldmedal = gold_medal.Team.value_counts().reset_index(name = 'Medal').head(5)
g = sns.catplot(x = "index", y = "Medal", data = totalgoldmedal,
               height = 7, kind = "bar", palette = "rocket")
g.despine(left = True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title("gold medals per country")
```
