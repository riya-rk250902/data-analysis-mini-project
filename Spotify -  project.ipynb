```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
spotdata = pd.read_csv(r"C:\Users\Lenovo\csv data_files\csv file sample_data\doc_1\spotify-data1.csv")
spotdata.head(5)
```

```python
# top 10 songs with least key present on this dataset.
sorted_data = spotdata.sort_values('key', ascending = True).head(10)
sorted_data
```

```python
# descriptive statistics for the numerical data present in the column.
spotdata.describe().transpose()
```

```python
#[:10] slicing operator
highest_key = spotdata.query('key>4', inplace = False).sort_values('key', ascending = False)
highest_key[:10]
```

```python
spotdata.set_index('date_added', inplace = True)
spotdata.index = pd.to_datetime(spotdata.index)
spotdata.head()
```

```python
spotdata[['first_artist']].iloc[4]
```

```python
# in this table we have duration but in miliseconds so we are going to convert it in seconds.
spotdata['duration'] = spotdata['duration_ms'].apply(lambda x: round(x/1000))
spotdata.head(5)
# spotdata.drop('duration_ms', inplace = True, axis = 1)
```

```python
spotdata.duration.head()
```

```python
# visualisation which is a correlation map.
# annot  =  annotation, fmt = string formatting, ticklabels is a matplotlib function
corr_data = spotdata.drop(['neg','key','mode'],axis =1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_data, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "Spectral", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
# .4 percent of my total dataset is: 
sample_spot = spotdata.sample(int(0.004*len(spotdata)))
```

```python
# 2nd dataset 
spotdata_genre = pd.read_csv(r"C:\Users\Lenovo\csv data_files\csv file sample_data\doc_1\spotify _genre _ data2.csv")
```

```python
# liveness of the songs for different songs.
plt.title("liveness of the songs for different songs")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y= 'the genre of the track', x ='Liveness - The higher the value, the more likely the song is a live recording', data = spotdata_genre)
plt.xlabel("Liveness - The higher the value, the more likely the song is a live recording")
plt.ylabel("the genre of the track")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous = spotdata_genre.sort_values("Popularity- The higher the value the more popular the song is", ascending = False).head(10)
sns.barplot(y = 'the genre of the track', x = 'Popularity- The higher the value the more popular the song is', data = famous).set(title = "top 5 genre by popularity")
```
