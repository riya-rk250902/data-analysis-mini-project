```python
# top 10 countries with most CO2 emmissions
# visualising countries data in terms of carbon emmission.
avg_co2_emission_acctoentity = df.groupby('Entity')['Value_co2_emissions_kt_by_country'].mean()
avg_co2_emission_acctoentity.head()

#highest_top5_countries =  Value_co2_emissions_kt_by_country.nlargest(10)
```

```python
top_values= avg_co2_emission_acctoentity.nlargest(5)
```

```python
# highest no of co2 emission 
plt.figure(figsize = (7,5))
sns.barplot(x = top_values.index, y = top_values.values)
plt.xlabel('Entity')
plt.ylabel('Average co2 emission')
plt.title('countries with highest number of co2 emissions')
plt.show()

# lowest no of co2 emission
plt.figure(figsize =(7,5))
lowest_country = avg_co2_emission_acctoentity.nsmallest(5)
sns.barplot(x = lowest_country.index, y = lowest_country.values)
plt.xlabel('Entity')
plt.ylabel('Average co2 emission')
plt.title('countries with the lowest rate of emissions')
plt.show()
```
