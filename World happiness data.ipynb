```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# runtime configuration settings.
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10,7)
plt.rcParams['figure.facecolor'] = 'pink'
```

```python
worlddata = pd.read_csv(r"C:\Users\Lenovo\csv data_files\csv file sample_data\doc_1\world_happiness_2021.csv")
worlddata.head(5)
```

```python
mydata = ['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support', 'Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
data = worlddata[mydata].copy()
```

```python
happy_data = data.rename({'Country name':'Country_name', 'Regional indicator': 'regional_indicator','Ladder score':'happiness score','Logged GDP per capita': 'lgdp per_capita','Social support': 'social_support','Healthy life expectancy': 'HLE','Freedom to make life choices':'FMLC','Generosity': 'generosity','Perceptions of corruption': 'corruption'})
```

```python
happy_data.head(20)
```

```python
happy_data.isnull().sum()
```

```python
# plot between happiness and gdp
plt.rcParams['figure.figsize'] = (15,7)
plt.title('plot b/w happiness and GDP score')
sns.scatterplot(x = 'happiness score', y = 'lgdp per_capita', hue = happy_data.regional_indicator, size = 200, data = happy_data);      
plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('happiness score')
plt.ylabel('lgdp per_capita')
```

```python
# plot pyplot to understand GDP by region.
# sum of all the countries for the different regions and there GDP all total
gdp_region = happy_data.groupby('regional_indicator')['lgdp per_capita'].sum()
gdp_region
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title("gdp by region")
plt.ylabel('')
```

```python
# now we are going to use the group by function that is part of pandas library and will count the total number of countries in each region.
total_country = happy_data.groupby('regional_indicator')[['Country_name']].count()
print(total_country)
```

```python
# correlation map 
# mask is a boolean array.
cor = happy_data.corr(method = "pearson")
f, ax= plt.subplots(figsize = (10,5))
sns.heatmap(cor, mask = np.zeros_like(cor, dtype = np.bool_),
cmap = "Blues", square = True, ax = ax )
```

```python
# corruption in regions 
corrupt = happy_data.groupby('regional_indicator')[['corruption']].mean()
corrupt
```

```python
# ha  = horizontal alignment
# we have made a bar chart of the table mentioned above: 
plt.rcParams['figure.figsize'] = (12,8)
plt.title('perception of corruption in various region')
plt.xlabel('Region', fontsize = 15)
plt.ylabel('corruption_index', fontsize = 15)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(corrupt.index, corrupt.corruption)
```

```python
# top 10 happiest countries and bottom 10 happiest countries
top_10 = happy_data.head(10)
bottom_10 = happy_data.head(10)
```

```python
# here we have given figure size and axes
fig, axes = plt.subplots(1,2, figsize = (16,6))
plt.tight_layout(pad = 2)
xlabels = top_10.Country_name
axes[0].set_title('top 10 happiest country life expectancy')
axes[0].set_xticklabels(xlabels, rotation = 45, ha = 'right')
sns.barplot(x = top_10.Country_name,  y = top_10.HLE, ax = axes[0])
axes[0].set_xlabel('Country_name')
axes[0].set_ylabel('life expectancy')

xlabels = bottom_10.Country_name
axes[0].set_title('top 10 countries with least life expectancy')
axes[0].set_xticklabels(xlabels, rotation = 45, ha = 'right')
sns.barplot(x = bottom_10.Country_name,  y = bottom_10.HLE, ax = axes[1])
axes[0].set_xlabel('Country_name')
axes[0].set_ylabel('life expectancy')
```

```python
# 'hue' is used to visualize the data of different categories in one plot.
plt.rcParams['figure.figsize'] = (15,7)
plt.title('plot b/w happiness and freedom')
sns.scatterplot(x = 'FMLC', y = 'happiness score', hue = happy_data.regional_indicator, size = 200, data = happy_data);      
plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('freedom')
plt.ylabel('happiness')
```

```python
# corrupt countries.
country = happy_data.sort_values(by = 'corruption').tail(10)
plt.rcParams['figure.figsize'] = (12,6)
plt.title('perception of the most corrupt countries in the world')
plt.xlabel('Country', fontsize = 15)
plt.ylabel('corruption_index', fontsize = 15)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.Country_name, country.corruption)
```

```python
# corruption varies in terms of happiness score.
plt.rcParams['figure.figsize'] = (15,7)
sns.scatterplot(x = 'happiness score', y = 'corruption', hue = happy_data.regional_indicator, size = 200, data = happy_data);  
plt.legend(loc = 'lower left', fontsize = '15')
plt.xlabel('happiness score')
plt.ylabel('corruption')
```
